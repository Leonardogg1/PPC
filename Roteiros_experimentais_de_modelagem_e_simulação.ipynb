#Questão 1

a) **Variável controlada:** A variável controlada é a grandeza ou parâmetro específico de um sistema que se deseja manter ou controlar em um determinado valor desejado. Em um sistema dinâmico, essa variável é influenciada pelo sinal de controle e por outros fatores, sendo o foco do controle para garantir que permaneça dentro de limites desejados ou siga um padrão desejado.

b) **Sinal de controle:** O sinal de controle é a entrada ajustável ou manipulável aplicada a um sistema dinâmico com o objetivo de influenciar o comportamento da variável controlada. Esse sinal é geralmente gerado pelo controlador e pode ser modificado para manter ou ajustar a variável controlada conforme necessário.

c) **Distúrbios:** Distúrbios são variações ou interferências não desejadas que afetam o comportamento de um sistema. Essas perturbações podem ser internas ou externas e têm o potencial de influenciar a variável controlada, tornando o controle mais desafiador.

d) **Controle feedback:** Controle feedback é uma abordagem em que a saída de um sistema é monitorada e comparada com a referência desejada. Com base nessa comparação, são feitos ajustes no sinal de controle para minimizar as diferenças entre a saída real e a saída desejada, contribuindo assim para o controle efetivo da variável controlada.

e) **Sistemas com malha aberta e com malha fechada:** Sistemas com malha aberta são sistemas em que o sinal de controle é aplicado sem considerar a saída do sistema. Em contraste, sistemas com malha fechada monitoram a saída do sistema e ajustam o sinal de controle com base nessa saída, permitindo uma resposta mais precisa e adaptativa às mudanças no sistema.

f) **Sistemas lineares:** Sistemas lineares são sistemas nos quais a relação entre as entradas e saídas pode ser representada por equações lineares. Isso implica que a resposta do sistema a uma entrada é proporcional à magnitude da entrada, e a superposição de diferentes entradas resulta em uma saída equivalente à soma das saídas individuais. Essa propriedade simplifica a análise e o projeto de sistemas em muitos casos.

g) **Função de Transferência de Sistemas lineares:** A função de transferência é uma ferramenta essencial na análise e projeto de sistemas lineares, proporcionando uma descrição matemática da relação entre a entrada e a saída de um sistema dinâmico. Ela é particularmente útil em sistemas lineares invariantes no tempo (LTI). Vamos discutir a função de transferência nos domínios temporal e de Laplace, sua obtenção e como ela é transferida entre esses domínios.

Forma nos Domínios Temporal e de Laplace:
Domínio Temporal:
A função de transferência no domínio temporal descreve a relação entre a entrada
$u(t)$ e a saída $y(t)$ de um sistema linear. Em sua forma geral, para um sistema linear de ordem $n$, a função de transferência é expressa como uma equação diferencial ordinária (EDO) de coeficientes constantes.

A forma geral no domínio temporal para um sistema de ordem $n$ é:

$
a_n \frac{d^n y(t)}{dt^n} + a_{n-1} \frac{d^{n-1} y(t)}{dt^{n-1}} + \ldots + a_1 \frac{dy(t)}{dt} + a_0 y(t) = b_m \frac{d^m u(t)}{dt^m} + b_{m-1} \frac{d^{m-1} u(t)}{dt^{m-1}} + \ldots + b_1 \frac{du(t)}{dt} + b_0 u(t)
$



Domínio de Laplace:
A transformada de Laplace é uma ferramenta matemática que permite transferir equações diferenciais no domínio temporal para o domínio de Laplace. No domínio de Laplace, a função de transferência é representada por uma expressão algébrica.

A forma geral da função de transferência no domínio de Laplace é:

 $
G(s) = \frac{Y(s)}{U(s)} = \frac{b_m s^m + b_{m-1} s^{m-1} + \ldots + b_1 s + b_0}{a_n s^n + a_{n-1} s^{n-1} + \ldots + a_1 s + a_0}
$

Obtenção da Função de Transferência:
A função de transferência pode ser obtida pela aplicação da transformada de Laplace às equações diferenciais que descrevem o sistema. A relação entre a entrada e a saída no domínio de Laplace é expressa pela razão das transformadas de Laplace de $y(t)$ e $u(t)$.

Transferência entre Domínios:
Domínio Temporal para Laplace:
A transformação de Laplace é aplicada à equação diferencial do sistema para obter a função de transferência.

Laplace para Domínio Temporal:
Para retornar ao domínio temporal a partir do domínio de Laplace, é realizada a transformada inversa de Laplace, que converte a função de transferência de volta para a forma temporal.

Forma Genérica:
A função de transferência genérica de um sistema linear é expressa como a razão polinomial de polinômios no numerador e no denominador, representando as contribuições da entrada e do sistema, respetivamente.

$G(s)=\frac{D(s)}{N(s)}​$


Onde:

$N(s)$ é um polinômio em $s$ que representa a contribuição da entrada.

$D(s)$ é um polinômio em $s$ que representa a resposta do sistema.
A forma genérica fornece uma maneira concisa de representar a função de transferência e facilita a análise e o projeto de sistemas lineares.

a) Sistema de Primeira Ordem com Entrada Degrau:

Um exemplo prático de um sistema de primeira ordem recebendo uma entrada degrau pode ser encontrado em um circuito de carga e descarga de um capacitor. Considere um capacitor inicialmente descarregado conectado a uma fonte de alimentação por meio de um resistor. Quando a fonte é ligada, a tensão no capacitor aumenta gradualmente de acordo com a constante de tempo do circuito RC (produto da resistência e capacitância). O comportamento dinâmico desse sistema pode ser modelado por uma equação diferencial de primeira ordem.

b) Sistema de Primeira Ordem com Entrada Rampa:

Um exemplo prático de um sistema de primeira ordem recebendo uma entrada rampa é um tanque de armazenamento de líquidos com uma entrada de fluxo constante. Se o fluxo de entrada for mantido constante, a altura da água no tanque aumentará linearmente ao longo do tempo. O comportamento dinâmico desse sistema pode ser descrito por uma equação diferencial de primeira ordem.

c) Sistema de Segunda Ordem com Entrada Impulso:

Considere um sistema massa-mola-amortecedor, como um carro suspenso sobre molas e amortecedores. Se o sistema estiver em repouso e for submetido a um impulso inicial, como uma perturbação rápida na estrada, o comportamento dinâmico do sistema será modelado por uma equação diferencial de segunda ordem. O impulso inicial pode representar, por exemplo, uma colisão súbita.

d) Sistema de Segunda Ordem com Entrada Degrau:

Um exemplo prático de um sistema de segunda ordem recebendo uma entrada degrau é um sistema de controle de temperatura em um ambiente fechado. Quando há uma mudança repentina na temperatura ambiente (entrada degrau), o sistema de controle de temperatura, que pode incluir um sensor de temperatura, um controlador e um sistema de aquecimento/arrefecimento, responderá com um comportamento dinâmico que pode ser modelado por uma equação diferencial de segunda ordem.








```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def V0(t):
    return 0.0 if t < 0.5 else 10.0

def mechanical_system(t, y):
    x, v, i = y
    m = 0.5
    b = 1
    k = 60
    a = 2.5e-2
    L_prime = 46.8e-3
    L = L_prime * (a / (a + x))
    dL_dx = a * L_prime / (a + x)**2
    F_ext = (i**2 / 2) * dL_dx  # Updated external force

    dvdt = (F_ext - b*v - k*(x-0.05)) / m
    dxdt = v

    # Electrical equation: di/dt = V0 - Ri - i(dL/dx)(dx/dt)
    R = 5
    didt = (V0(t) - R * i - i * dL_dx * v) * ((a + x) / x)

    return [dxdt, dvdt, didt]

# Initial conditions
x0 = 0.05
v0 = 0.0
i0 = 0.0
y0 = [x0, v0, i0]
T = 4
dt = 0.0071428571
t_span = (0, T)

# Solve the ODE system using solve_ivp
sol = solve_ivp(mechanical_system, t_span, y0, method='RK45', t_eval=np.arange(0, T, dt), dense_output=True)

# Extract the solution
t_mechanical = sol.t
x_mechanical = sol.y[0]
v_mechanical = sol.y[1]
i_mechanical = sol.y[2]

# Calculate magnetic force
a = 2.5e-2
L_prime = 46.8e-3
dL_dx = a * L_prime / (a + x_mechanical)**2
force_magnetic = (i_mechanical**2 / 2) * dL_dx

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t_mechanical, x_mechanical, label='Displacement (m)')
plt.title('Mechanical System SSO')
plt.xlabel('Time (s)')
plt.legend()
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(t_mechanical, i_mechanical, label='Current (A)')
plt.title('Electrical System SPO')
plt.xlabel('Time (s)')
plt.legend()
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(t_mechanical, force_magnetic, label='Magnetic Force (N)')
plt.title('Magnetic Force')
plt.xlabel('Time (s)')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Constantes
R = 8.314  # Constante universal dos gases (J/mol*K)
Ea = 50000  # Energia de ativação (J/mol)
k0 = 1e6  # Fator pre-exponencial (1/s)

# EDO do sistema hidráulico para o primeiro tanque
def hydraulic_system_tank1(t, h1, q_in, A_tank1):
    C_d = 0.6  # Coeficiente de descarga
    g = 9.81  # Aceleração da gravidade (m/s^2)
    q_out1 = C_d * A_tank1 * np.sqrt(2 * g * h1)  # Vazão de saída (m^3/s)
    dh1_dt = (q_in - q_out1) / A_tank1  # Taxa de variação do nível de fluido no primeiro tanque
    return dh1_dt, q_out1

# EDO do sistema químico para o primeiro tanque
def chemical_reaction_tank1(t, C_A1, k):
    dC_A1_dt = -k * C_A1  # Taxa de variação da concentração de A no primeiro tanque
    return dC_A1_dt

# EDO do sistema hidráulico para o segundo tanque
def hydraulic_system_tank2(t, h2, q_out1, A_tank2):
    C_d = 0.6  # Coeficiente de descarga
    g = 9.81  # Aceleração da gravidade (m/s^2)
    q_out2 = C_d * A_tank2 * np.sqrt(2 * g * h2)  # Vazão de saída (m^3/s)
    dh2_dt = (q_out1 - q_out2) / A_tank2  # Taxa de variação do nível de fluido no segundo tanque
    return dh2_dt

# EDO do sistema químico para o segundo tanque com aquecimento
def chemical_reaction_tank2(t, C_A2, T2):
    k_T2 = k0 * np.exp(-Ea / (R * T2))  # Constante de reação Arrhenius
    dC_A2_dt = -k_T2 * C_A2  # Taxa de variação da concentração de A no segundo tanque
    return dC_A2_dt

# EDO do sistema térmico para o segundo tanque
def thermal_system_tank2(t, T2, q_out1, T_in, C_p, m):
    U = 100  # Coeficiente de transferência de calor (W/m^2*K)
    A = 1.0  # Área de troca térmica (m^2)
    T_ext = 373  # Temperatura externa ou da fonte de calor (K)

    # Equação de balanço de energia
    dT2_dt = (U * A * (T_ext - T2) + q_out1 * C_p * (T_in - T2)) / (m * C_p)
    return dT2_dt

# Função para resolver o sistema acoplado
def system_coupled(t, y, k, q_in, A_tank1, A_tank2, T_in, C_p, m):
    C_A1, h1, C_A2, h2, T2 = y

    # Reações químicas e sistema hidráulico no primeiro tanque
    dC_A1_dt = chemical_reaction_tank1(t, C_A1, k)
    dh1_dt, q_out1 = hydraulic_system_tank1(t, h1, q_in, A_tank1)

    # Reações químicas, sistema hidráulico e térmico no segundo tanque
    dC_A2_dt = chemical_reaction_tank2(t, C_A2, T2)
    dh2_dt = hydraulic_system_tank2(t, h2, q_out1, A_tank2)
    dT2_dt = thermal_system_tank2(t, T2, q_out1, T_in, C_p, m)

    return [dC_A1_dt, dh1_dt, dC_A2_dt, dh2_dt, dT2_dt]

# Condições iniciais
C_A1_0 = 1.0  # Concentração inicial de A no primeiro tanque (mol/L)
h1_0 = 1.0    # Nível inicial do fluido no primeiro tanque (m)
C_A2_0 = 0.0  # Concentração inicial de A no segundo tanque (mol/L)
h2_0 = 0.5    # Nível inicial do fluido no segundo tanque (m)
T2_0 = 300.0  # Temperatura inicial no segundo tanque (K)
y0 = [C_A1_0, h1_0, C_A2_0, h2_0, T2_0]

# Parâmetros do sistema
k = 0.1  # Constante da reação no primeiro tanque (1/s)
q_in = 0.01  # Vazão de entrada no primeiro tanque (m^3/s)
A_tank1 = 1.0  # Área da seção transversal do primeiro tanque (m^2)
A_tank2 = 1.5  # Área da seção transversal do segundo tanque (m^2)
T_in = 300.0  # Temperatura de entrada do fluido no segundo tanque (K)
C_p = 4184  # Capacidade térmica específica da água (J/kg*K)
m = 100.0  # Massa do fluido no segundo tanque (kg)

# Intervalo de tempo
t_span = (0, 2000)  # Tempo total de simulação (s)
t_eval = np.linspace(0, 2000, 1000)  # Pontos onde a solução será avaliada

# Resolver o sistema usando solve_ivp com RK45
solution = solve_ivp(system_coupled, t_span, y0, args=(k, q_in, A_tank1, A_tank2, T_in, C_p, m), method='RK45', t_eval=t_eval)

# Extrair soluções
t = solution.t
C_A1 = solution.y[0]
h1 = solution.y[1]
C_A2 = solution.y[2]
h2 = solution.y[3]
T2 = solution.y[4]

# Plotar os resultados
plt.figure(figsize=(14, 10))

plt.subplot(3, 1, 1)
plt.plot(t, C_A1, label='Concentração de A no Tanque 1 (mol/L)')
plt.plot(t, C_A2, label='Concentração de A no Tanque 2 (mol/L)', linestyle='--')
plt.title('Sistema Químico')
plt.xlabel('Tempo (s)')
plt.ylabel('Concentração (mol/L)')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, h1, label='Nível de Fluido no Tanque 1 (m)')
plt.plot(t, h2, label='Nível de Fluido no Tanque 2 (m)', linestyle='--')
plt.title('Sistema Hidráulico')
plt.xlabel('Tempo (s)')
plt.ylabel('Nível (m)')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, T2, label='Temperatura no Tanque 2 (K)', color='red')
plt.title('Sistema Térmico no Tanque 2')
plt.xlabel('Tempo (s)')
plt.ylabel('Temperatura (K)')
plt.legend()

plt.tight_layout()
plt.show()

# Calcular parâmetros de desempenho para o segundo tanque
Tp2, Ts2, Mp2 = calculate_performance_parameters(t, C_A2)

print(f'Tempo de Pico no Tanque 2 (Tp2): {Tp2:.2f} s')
print(f'Tempo de Acomodação no Tanque 2 (Ts2): {Ts2:.2f} s')
print(f'Overshoot no Tanque 2 (Mp2): {Mp2:.2f}')
```
